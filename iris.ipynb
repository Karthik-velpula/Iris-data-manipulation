```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
data=load_iris()
df=pd.DataFrame(data.data,columns=data.feature_names)
df.head()
```

```python
df.info()
```

```python
print(df.info())
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
df.iloc[0,0]=None
df.iloc[2,1]=None
df.iloc[3,2]=None
print(df.isnull().sum())
```

```python
df.fillna(df.mean(),inplace=True)
print(df.isnull().sum())
```

```python
df.hist(figsize=(10,10),bins=20,color='orange',edgecolor='black')
plt.suptitle('Histogram of Iris Dataset')
plt.tight_layout()
plt.show()
```

```python
df.fillna(df.median(),inplace=True)
print(df.isnull().sum())
```

```python
df.fillna(df.mode(),inplace=True)
print(df.isnull().sum())
```
